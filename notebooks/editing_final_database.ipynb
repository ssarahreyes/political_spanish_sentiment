```python
def madrid_campaña(df, column):
    
    column_list = list(column)
    
    campaña = []
    
    for t in column_list:
        if 'Madrid' in t:
            campaña.append('madrid')
        else:
            campaña.append('general')
            
    df['campaña'] = campaña
    
    return df

# pasarle esta función a todo el datafrae que tenga mañana y añadirla en las funciones del main script.
```
