```python
def madrid_campa√±a(df, column):
    
    column_list = list(column)
    
    campa√±a = []
    
    for t in column_list:
        if 'Madrid' in t:
            campa√±a.append('madrid')
        else:
            campa√±a.append('general')
            
    df['campa√±a'] = campa√±a
    
    return df

# pasarle esta funci√≥n a todo el datafrae que tenga ma√±ana y a√±adirla en las funciones del main script.
```
